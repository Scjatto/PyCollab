## **Text To Speech Block**

### All Possible languages and Accents
 

```python
tts_en = gTTS('HELLO, this is Shizuku-chan, welcome to my channel', lang='en')
tts_fr = gTTS('bonjour', lang='ja')

with open('OutputFiles/hello_bonjour.mp3', 'wb') as f:
    tts_en.write_to_fp(f)
    tts_fr.write_to_fp(f)
# tts.save('OutputFiles/hello.mp3')
# mp3_fp = BytesIO()
# tts_en.write_to_fp(mp3_fp)
```
